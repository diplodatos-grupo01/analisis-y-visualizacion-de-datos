IMPORTANTE: ¿Qué hacemos con los datos faltantes?

```python
# TODOOOOOOOOOOO: usar bien los isnan. value_counts, que se uso en todas las funciones! tiene un param isnan
def per_year_events():
    yeardataset = dataset[(dataset.year != 0 )].sort_values('year', ascending=False).year.value_counts()
    ax = seaborn.barplot(data=yeardataset.to_frame().reset_index(),
        y='year', x='index', color='#2ecc71', orient="v")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
    ax.set_xlabel('año')
    ax.set_title('Eventos de violencia por año')
    seaborn.despine()
per_year_events()
```

```python
def per_year_freq_probabilities():
    yeardataset = dataset[(dataset.year != 0 )].sort_values('year', ascending=False).year.value_counts(normalize=True)
    ax = seaborn.barplot(data=yeardataset.to_frame().reset_index(),
        y='year', x='index', color='#2ecc71', orient="v")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
    ax.set_ylabel('probabilidad')
    ax.set_xlabel('año')
    ax.set_title('Probabilidades de eventos por año')
    seaborn.despine()
per_year_freq_probabilities()
```

```python
def per_year_sums_probability():
    yeardataf = dataset[(dataset.year != 0 )].year.value_counts(normalize=True).to_frame().reset_index()
    yeardf = yeardataf.sort_values('index')
    yeardf['sums'] = yeardf['year'].cumsum()
    ax = seaborn.barplot(data=yeardf,
        y='sums', x='index', color='#2ecc71', orient="v")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
    ax.set_title('Probabilidad acumulada de eventos por año')
    seaborn.despine()
per_year_sums_probability()
```
